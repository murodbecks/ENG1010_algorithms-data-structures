You are given a list of positive numbers and a target value. Your task is to examine the list and determine the smallest contiguous section of the list whose elements add up to at least the target value. What “contiguous” means: The elements must appear next to each other in the list. You are not allowed to reorder the list or skip elements.

```
Example
[2, 3, 1, 2, 4, 3]
target=7
Some possible contiguous groups are:
[2, 3, 1, 2] → sum = 8 (length 4)
[1, 2, 4] → sum = 7 (length 3)
[4, 3] → sum = 7 (length 2)
Among all valid choices, [4, 3] is the shortest, so the answer is 2.
```

Provide a solution to this puzzle and comment about its complexity. Then, develop a Python program to solve it. Experiment with different list sizes and different target values. You may use random numbers to populate the list.

````python
# Complexity: O(2*n) = O(n)
# Memory: O(n)

def get_contigious_list(lst: list, target: int):
    def is_valid_list(given_list: list, target: int = target):
        if not given_list:
            return False
        
        return sum(given_list) >= target
    
    start_idx = 0
    end_idx = 0
    
    contigious_list = []
    
    while not is_valid_list(lst[start_idx:end_idx]) and end_idx < len(lst):
        end_idx += 1

        if end_idx == len(lst) and (not is_valid_list(lst[start_idx:end_idx])):
            return contigious_list

    contigious_list = lst[start_idx:end_idx]
    
    while end_idx <= len(lst):
        start_idx += 1
    
        if is_valid_list(lst[start_idx:end_idx]):
            if len(contigious_list) > start_idx - end_idx + 1:
                contigious_list = lst[start_idx:end_idx]
        else:
            end_idx += 1

    return contigious_list

lst = [2, 3, 1, 2, 4, 3]
target = 7
contigious_list = get_contigious_list(lst, target)
print(contigious_list, len(contigious_list))
````

# week2

Given a list of n distinct integers and a sequence of n boxes with preset inequality signs inserted between them, design an algorithm that places the numbers into the boxes to satisfy those inequalities. For example, the numbers 2, 5, 1, and 0 can be placed in the four boxes as shown below:

```
___ < ___ > ___ < ___

0 < 5 > 1 < 2
```

Provide a solution to this puzzle and comment about its complexity. Then, develop a Python program to solve it. Experiment with different numbers. You may use random numbers to populate the list and generate the inequalities.

````python
# Complexity: O(2*nlogn + 2n) = O(nlogn)
# Memory: O(5n+3) = O(n)

def sort_with_inequalities(numbers: list, inequality_signs: str) -> str:
    # finding output ranks
    output_ranks = []
    
    current_rank = 0
    output_ranks.append(current_rank)
    for sign in inequality_signs:
        if sign == "<":
            current_rank += 1
        elif sign == ">":
            current_rank -= 1
        else:
            raise ValueError(f"Unexpected Value: {sign} should be one of `<` or `>`")
    
        output_ranks.append(current_rank)
    
    # mapping output ranks to numbers
    rank_to_number_map = {}
    for num, rank in zip(sorted(numbers), sorted(output_ranks)):
        if rank not in rank_to_number_map:
            rank_to_number_map[rank] = []
    
        rank_to_number_map[rank].append(num)
    
    # generating output from mapping
    output = []
    for rank in output_ranks:
        output.append(rank_to_number_map[rank].pop())

    # generating output string
    output_str = str(output[0])
    for sign, num in zip(inequality_signs, output[1:]):
        output_str += f" {sign} {num}"
    
    return output_str

numbers = [2, 5, 1, 0, 7]
inequality_signs = "><<>"
sort_with_inequalities(numbers, inequality_signs)
````

# week3

You are given a building with N floors and exactly two identical eggs, and there exists an unknown highest floor F such that an egg dropped from any floor higher than F will break, while an egg dropped from floor F or below will not. A broken egg can no longer be used, while an egg that does not break can be used again, multiple times until it breaks. The task is to determine the value of F by dropping the eggs, subject to these rules, while minimizing the maximum number of drops required in the worst case. The difficulty of the problem comes from the fact that each drop provides limited information and breaking an egg reduces the resources available for future tests.

You should provide at least two algorithmic solutions to this problem: a baseline solution, and a more efficient one. Recall that your task is to minimize the number of drops in the worst case, providing the optimal solution

````python
# Naive solution - check all floors from bottom until the egg breaks
# here we can only check whether 

# Time Complexity: O(F)
# Space Complexity: O(1)

def breaking_eggs_naive(n, f, start=0):
    num_steps = 0
    
    for i in range(start, n+1):
        num_steps += 1
        if i > f:
            return i-1, num_steps

    return -1, num_steps

breaking_eggs_naive(100, 45)
````

````python
# Optimized solution - we do binary search until our first egg breaks then we go back to naive solution with limited search space

# Time Complexity: O(N)
# Space Complexity: O(1)

def breaking_eggs_optimized(n, f):
    num_steps = 0
    
    low = 0
    high = n

    while low < high:
        num_steps += 1
        
        mid = (low+high)//2
        
        if mid > f:
            i, num_steps1 = breaking_eggs_naive(high, f, start=low)
            return i, num_steps + num_steps1
        else:
            low = mid

    return -1, num_steps

breaking_eggs_optimized(100, 55)
````

# week4

A museum originally had n + 1 ancient plaques, numbered consecutively from 0 to n. One plaque has been stolen. The remaining n plaques are still mounted in increasing order along a hallway, and the curator records their numbers into an array A[0..n − 1]. Exactly one number from 0 to n is missing. Your task is to help the curator find the missing plaque as quickly as possible.

```
Example
A = [0, 1, 2, 3, 5, 6, 7]
Missing plaque: 4
```

````python
import random

n = 10

numbers = list(range(n+1))

# random.seed(25)
missing = random.choice(numbers)
numbers.remove(missing)

def find_missing(numbers):
    n = len(numbers)
    low = 0
    high = n - 1
    
    while low <= high:
        mid = (low + high) // 2
        
        if numbers[mid] == mid:
            # Missing number is to the right
            low = mid + 1
        else:
            # Missing number is at mid or to the left
            high = mid - 1
    
    # When loop exits, low is the missing number
    return low

find_missing(numbers), missing
````

# week5

Indiana Jones (the most famous archaeologist) is excavating a lost temple. Each artifact that he uncovers has a danger level (i.e., a number indicating how unstable or cursed it is). After every discovery, Indy must immediately know the median danger level so he can decide whether to proceed the excavation or run!.
- Artifacts appear one at a time (when a new artifact is available, the median must be updated)
- The temple reacts if Indy spends too long analyzing (the solution must be efficient)
- Sorting/Re-sorting artifacts after each find would trigger traps (resorting is not allowed).
- If the number of artifacts is even, the median is computed as the average of the two ”middle” artifacts.

Example
```
5, 15, 1, 3, 8, 7, 9
new artifact 5: median = 5
new artifact 15: median = (5 + 15)/2 = 10
new artifact 1: median = 5
new artifact 3: median = (5 + 3)/2 = 4
...
```

````python
# here we use 2 heap min-priority queues to track the values that are:
# - the largest value that is smaller or equal to the median
# - the smallest value that is larger than the median

# inserting a number: O(logn) - heappush and rebalancing
# getting a median: O(1)

import heapq

class MedianFinder():
    def __init__(self):
        # for storing larger half
        self.min_heap = []
        # for storing lower half 
        self.max_heap = []

    def insert_num(self, num):
        if not self.max_heap or num <= -self.max_heap[0]:
            heapq.heappush(self.max_heap, -num)
        else:
            heapq.heappush(self.min_heap, num)

        self._rebalance()

    def _rebalance(self):
        if len(self.max_heap) - len(self.min_heap) > 1:
            val = -heapq.heappop(self.max_heap)
            heapq.heappush(self.min_heap, val)
        elif len(self.min_heap) > len(self.max_heap):
            val = heapq.heappop(self.min_heap)
            heapq.heappush(self.max_heap, -val)

    def get_median(self):
        if len(self.max_heap) == len(self.min_heap):
            return (-self.max_heap[0]+self.min_heap[0])/2
        else:
            return -self.max_heap[0]

def find_running_median(artifacts):
    median_finder = MedianFinder()

    for i, artifact in enumerate(artifacts):
        median_finder.insert_num(artifact)
        median = median_finder.get_median()

        print(f"new artifact {artifact}: median = {median}")

find_running_median([5, 15, 1, 3, 8, 7, 9])
````

An ancient structure is organized as an N -ary tree, where each node contains a positive numerical value. Legend says that the path to the inner sanctum is revealed only when a specific total sum, K, is found within the structure’s branches.
Your task is to determine if a path exists that sums exactly to K for five increasingly complex
scenarios:
- V1 (Root-to-Leaf ): The path must start at the root and terminate at a leaf node.
- V2 (Root-to-Anywhere): The path must start at the root but can end at any node along the downward descent.
- V3 (Anywhere-to-Anywhere Downward): The path can start at any node and must move downward to any of its descendants.
- V4 (The Universal Path): The path can start at any node and end at any other node in the tree.

Example
Consider a tree where the root is 5, with children 4 and 8. Node 4 has a child 11.
```
   5
  / \
  4 8
 /
11
```

If K = 20:
- V1 to V4: True (5 -> 4 -> 11)

If K = 9:
- V1: False (No root-to-leaf path sums to 9)
- V2-V3: True (5 -> 4)

If K = 15:
- V1: False (No root-to-leaf path sums to 15)
- V2: False (No root-to-anywhere path sums to 15)
- V3-V4: True (4 -> 11)

If K = 17:
- V1: False (No root-to-leaf path sums to 17)
- V2: False (No root-to-anywhere path sums to 17)
- V3: False (No any downward path sums to 17)
- V4: True (4 -> 5 -> 8)

Notes
- A ”contiguous” path means you cannot skip nodes; every node in the sequence must be connected.
- In V4, the path is not restricted to a single downward direction; it can go up from a leaf toward the root and back down into a different subtree.

````python
class Node:
    def __init__(self, val):
        self.val = val
        self.children = []

    def add_child(self, child):
        self.children.append(child)

def print_tree(node, prefix="", is_last=True, is_root=True):
    if not node:
        return

    # Print the current node with the correct connector
    if is_root:
        print(node.val)
        new_prefix = ""
    else:
        connector = "└── " if is_last else "├── "
        print(prefix + connector + str(node.val))
        # Update the prefix for the children
        new_prefix = prefix + ("    " if is_last else "│   ")

    # Recursively call for children
    for i, child in enumerate(node.children):
        last_child = (i == len(node.children) - 1)
        print_tree(child, new_prefix, last_child, False)

root = Node(5)
root.add_child(Node(4))
root.add_child(Node(8))
root.children[0].add_child(Node(11))
root.children[0].children[0].add_child(Node(19))
root.children[0].children[0].add_child(Node(1))

print_tree(root)
````

````python
# Time Complexity: O(n)
# Space Complexity: O(h)

def solution_v1(root, k, current_sum=0):
    if not root:
        return False
        
    if current_sum > k:
        return False

    current_sum += root.val
    
    if len(root.children) == 0:
        return current_sum == k

    for child in root.children:            
        if solution_v1(child, k, current_sum):
            return True

    return False

solution_v1(root, 13), solution_v1(root, 16), solution_v1(root, 20), solution_v1(root, 21), solution_v1(root, 30)
````

````python
# Time Complexity: O(n)
# Space Complexity: O(h)

def solution_v2(root, k, current_sum=0):
    if not root:
        return False
    
    current_sum += root.val
    if current_sum > k:
        return False
    elif current_sum == k:
        return True

    for child in root.children:
        if solution_v2(child, k, current_sum):
            return True

    return False

solution_v2(root, 13), solution_v2(root, 16), solution_v2(root, 20), solution_v2(root, 21), solution_v1(root, 30)
````

````python
# Time Complexity:
# Space Complexity: O(n*h)

def solution_v3(root, k, vals=[]):
    if not root:
        return False

    for i in range(len(vals)):
        if sum(vals[i:]) + root.val == k:
            return True

    if len(root.children) == 0:
        return False

    for child in root.children:
        if solution_v3(child, k, vals+[root.val]):
            return True

    return False

solution_v3(root, 13), solution_v3(root, 20), solution_v3(root, 16), solution_v3(root, 21), solution_v3(root, 30)
````

# week7

In the year 3275, astronomers are mapping a massive 3-dimensional nebula. The nebula is divided into a cubic grid:
```
x = 0...10^9
y = 0...10^9
z = 0...10^9
```

In theory, the nebula contains: $(10^9)^3$ = 10^27 possible voxels (3D cells). Each voxel can be:
- 0: empty space
- x: high-energy plasma value ($x\neq0$)

But after decades of scanning, they detected only 300,000 active plasma voxels. Find an efficient way to represent the voxels and give all the details of your approach. Your solution must support inserting new non-empty voxels, deleting existing voxels if needed. Compare your solution to the baseline.

| Approach | Space | Lookup | Insert | Delete | Feasible? |
|----------|-------|--------|--------|--------|-----------|
| Full 3D array | O(10^27) | O(1) | O(1) | O(1) | Impossible |
| Sorted list of coordinates | O(n) | O(log n) | O(n) | O(n) | Yes, but slower inserts |
| **Hash table** | O(n) | O(1) avg | O(1) avg | O(1) avg | Yes |

````python
# we will use hash function (Hash with Chaining) to transform 3D coordinate into unique integer

# Time complexity (insertion, delition, query): O(n) in worst case, but O(1) in average case
# Space complexity: O(N)

PRIME_NUMBER = 10**9+7

class HashWithChaining():
    def __init__(self, size=300_001, prime_number=PRIME_NUMBER, a=250100, b=1235, c=90123):
        self.size = size
        self.prime_number = prime_number
        self.a = a
        self.b = b
        self.c = c
        
        self.hash_table = [None] * self.size

    def hash_function(self, val, initial_repr=0):
        x, y, z = val
        return (self.a * x + self.b * y + self.c * z) % self.prime_number % self.size

    def add_value(self, val):
        if self.has_value(val):
            return True
            
        hash_value = self.hash_function(val)
        
        if not self.hash_table[hash_value]:
            self.hash_table[hash_value] = {val}
            return True
        else:
            self.hash_table[hash_value].add(val)
            return True

    def has_value(self, val):
        hash_value = self.hash_function(val)
        bucket = self.hash_table[hash_value]
        
        if bucket is None:
            return False
        
        return val in bucket

    def delete_value(self, val):
        hash_value = self.hash_function(val)
        if  val in self.hash_table[hash_value]:
            self.hash_table[hash_value].discard(val)
            if len(self.hash_table[hash_value]) == 0:
                self.hash_table[hash_value] = None
            return True

        return False

ht = HashWithChaining()
ht.add_value((3, 4, 5))
ht.has_value((3, 4, 5))
````

You have n gas stations in a circle. At station i you gain gas[i] liters and it costs cost[i] liters to travel to the next station. You start with an empty tank.
Objective: Find the index of a station from which you can complete the full circle exactly once. If this is impossible, then return -1.

Example
gas = [1, 2, 3, 4, 5]
cost = [3, 4, 5, 1, 2]
Answer: 3

Because starting at station 3:
```
tank = 0 + 4 - 1 = 3
tank = 3 + 5 - 2 = 6
tank = 6 + 1 - 3 = 4
tank = 4 + 2 - 4 = 2
tank = 2 + 3 - 5 = 0
```
You complete the circle.

````python
# we will go through gains list (gas-cost) and increment the starting index when tank goes below zero. 

# Time Complexity: O(n)
# Space Complexity: O(1)

def circular_tour(gas, cost):
    if sum(gas) < sum(cost):
        return -1
    
    start_idx = 0
    current_tank = 0
    
    for i in range(len(gas)):
        current_tank += gas[i] - cost[i]
        
        if current_tank < 0:
            start_idx = i + 1
            current_tank = 0
            
    return start_idx

print(circular_tour([1, 2, 3, 4, 5], [3, 4, 5, 1, 2])) # Output: 3
````
